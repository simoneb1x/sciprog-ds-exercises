## load

Write a function that given a `filename_prefix` and list of `languages`, parses the corresponding files and RETURNS a **dictionary of dictionaries**, which maps movies id to movies data, in the format as in the exerpt.

- When a label is missing, you will find instead an id like `Q3778078`: substitute it with empty string
 (HINT: to recognize ids you might use `is_digit()` method) 
- convert date numbers to proper integers
- **DO NOT** put constant ids nor language tags in the code (so no `'Q221074'` nor `'it'` ...)

```python
import csv
from pprint import pprint

def load(filename_prefix, languages):
    
    # inizializzo il dizionario main che conterrà i film, catalogati per ID
    main_dict = {}
    
    # apro il file -> funzione parziale perché il prefisso è statico
    with open('bud-spencer-terence-hill-movies-it.csv', encoding='utf-8', newline='') as f:
        my_reader = csv.reader(f, delimiter=",")
        
        # assegno la lista di my_reader ad una variabile per comodità
        my_reader_list = list(my_reader)
        
        # avvio un ciclo for per leggere tutte le righe del reader
        for row in my_reader_list:
            
            # skippo l'header
            if my_reader_list.index(row) == 0:
                continue
            
            # assegno il movie ID alla variabile
            movie_id = row[2][31:]
            
            # aggiungo l'ID del film al dizionario main. l'ID del film avrà un suo 
            # sottodizionario che conterrà tutti i suoi dettagli.
            main_dict[movie_id] = {
                'actors': [],
                'first_release': (int(row[4][0:4]), int(row[4][5:7]), int(row[4][8:10])),
                'names': {}
            }
    
            # pprint(row)
        
        # pprint(my_reader_list)
    
    with open('bud-spencer-terence-hill-movies-it.csv', encoding='utf-8', newline='') as f:
        my_reader = csv.reader(f, delimiter=",")
        
        # assegno la lista di my_reader ad una variabile per comodità
        my_reader_list = list(my_reader)
        
        # avvio un ciclo for per riempire i sottodizionari del main
        for row in my_reader_list:
            # skippo l'header
            if my_reader_list.index(row) == 0:
                continue
            
            # assegno il movie ID alla variabile
            movie_id = row[2][31:]
            
            # avvio un ciclo for annidato per aggiungere gli attori
            for row in my_reader_list:
                # se l'id del film della row letta corrisponde al movie_id principale, 
                # allora ne aggiungo tutti i dettagli
                if row[2][31:] == movie_id:
                    if main_dict[movie_id]['actors'].count((row[0][31:], row[1])) < 1:
                        main_dict[movie_id]['actors'].append((row[0][31:], row[1]))
       
    # comincio ad aggiungere tutti i vari nomi in diverse lingue   
    for lang in languages:
        with open(filename_prefix + '-' + lang + '.csv', encoding='utf-8', newline='') as f:
            my_reader = csv.reader(f, delimiter=",")
        
            # assegno la lista di my_reader ad una variabile per comodità
            my_reader_list = list(my_reader)
        
            # avvio un ciclo for per leggere tutte le righe del reader
            for row in my_reader_list:
                # skippo l'header
                if my_reader_list.index(row) == 0:
                    continue
                
                # assegno il movie ID alla variabile
                movie_id = row[2][31:]
                
                # assegno i nomi
                for elem in main_dict:
                    if movie_id == elem:
                        if row[3][1:].isdigit() == True:
                            main_dict[movie_id]['names'][lang] = ""
                        else:
                            main_dict[movie_id]['names'][lang] = row[3]
    
    # pprint(main_dict)
    return main_dict
    
movies_db = load('bud-spencer-terence-hill-movies', ['en', 'it', 'de'])
#movies_db = load('bud-spencer-terence-hill-movies', ['es', 'en', 'de','it'])
```

## save_table

Write a function that given a movies db and a list of `languages`, writes a new file `merged.csv` 

- separate actor names with `and`
- use only the year as date
- file must be formatted like this:

```python
import csv
from pprint import pprint

def save_table(movies, languages):
    with open('merged.csv', 'w', newline='') as csvfile_out:
        my_writer = csv.writer(csvfile_out, delimiter=',')
        
        header = ['movie_id'] + ['name ' + lan for lan in languages] + ['first_release', 'actors']
        my_writer.writerow(header)
        
        for movie_id in movies:
            row = [movie_id]
            list_of_actors = []
            # print(movies[movie_id])
            for language in languages:
                row.append(movies[movie_id]['names'][language])
            row.append(movies[movie_id]['first_release'][0])
            for actors in movies[movie_id]['actors']:
                list_of_actors.append(actors[1])
            if len(list_of_actors) >= 2:
                list_of_actors = ' and '.join(list_of_actors)
            else:
                list_of_actors = ''.join(list_of_actors)    
            row.append(list_of_actors)
            my_writer.writerow(row)
                
save_table(movies_db, ['en','it','de'])
#save_table(movies_db, ['de'])
```

## show_graph

Display a NetworkX graph of movies ([see examples](https://en.softpython.org/relational/relational1-intro-sol.html#Fancy-networkx-graphs)) from `since_year` (included) to  `until_year` (included), in the given `language`

* display actor names as capitalized (**DON'T** write constants in your code, so no `'Terence'` nor `'TERENCE'`...)
* display co-starred movies, non co-starred movies and actors with different colors by setting node attributes `style='filled'` and i.e. `fillcolor='green'` (see [some color names](https://www.w3.org/wiki/CSS/Properties/color/keywords))

```python
import networkx as nx
from sciprog import draw_nx
from pprint import pprint
import pandas as pd
import numpy as np
import csv

def show_graph(movies, since_year, until_year, language):
    
    G = nx.DiGraph()       
    G.graph['graph']= { 'layout':'neato'}  # don't delete these!
    
    new_list = []
    index = 0
    
    for elem in movies_db:
        #print(elem)
        
        if movies_db[elem]['first_release'][0] >= since_year and movies_db[elem]['first_release'][0] <= until_year:
            new_list.append([movies_db[elem]['names'][language]])
            for actors in movies_db[elem]['actors']:
                new_list[index].append(actors[1])
            index += 1
    
    list_actors = []
    
    for row in new_list:
        for elem in range(len(row)):
            if elem != 0:
                if list_actors.count(row[elem]) == 0:
                    list_actors.append(row[elem])
    
    index = 1
    
    couples_actors_id = []
    
    for actors in list_actors:
        G.add_node(index+1, label=actors, color='orange', fontcolor='black')
        couples_actors_id.append((index, actors))
        index += 1
    
    couples_film_id = []
    
    j = 0
    
    for row in new_list:
        index += 1
        for elem in range(len(row)):
            if elem == 0:
                G.add_node(index, label=row[elem], color='green', fontcolor='black')
                couples_film_id.append([index, row[elem]])
            for actors in couples_actors_id:
                if row[elem] == actors[1]:
                    couples_film_id[j].append(actors[0])
        j += 1
        
    
    for row in couples_film_id:
        for elem in row:
            if row.index(elem) == 0:
                film = elem
            if type(elem) != str and row.index(elem) > 0:
                G.add_edge(elem, film)
                
    
    print(couples_film_id)
    print(couples_actors_id)
    draw_nx(G)
    
    
        
    
show_graph(movies_db, 1970, 1975, 'en')    

```
